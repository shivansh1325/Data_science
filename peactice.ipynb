```python
colum_name ={
    'Rape': 'Rape Cases',
    'K&A': 'Kidnap and Assault',
    'DD': 'Dowry Deaths',
    'AoW': 'Assault on Women',
    'AoM': 'Assault on Minors',
    'DV': 'Domestic Violence',
    'WT': 'Witchcraft'
}
wocr.rename(columns= colum_name,inplace=True)
wocr.columns
```

```python
wocr.head()
```

```python
wocr[ ['Year', 'Rape Cases', 'Kidnap and Assault', 'Dowry Deaths',
       'Assault on Women', 'Assault on Minors', 'Domestic Violence',
       'Witchcraft']].corr()
```

```python
wocr['Year'].value_counts()
```

```python
wocr['Assault on Minors'].value_counts(bins=30) #average Assault on Minors crime in each state is between 0-1256
```

```python
wocr['Assault on Women'].value_counts(bins=15)
```

```python
crime_trends=wocr.groupby('Year').sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(12, 6))
sns.lineplot(data=crime_trends)
plt.title('Trend of Crimes Against Women (2001-2021)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Crime Type', loc='upper left')
plt.show()
```

```python
no_y_crime= wocr.drop(columns="Year")
```

```python
crime_by_state=no_y_crime.groupby("State").sum().sort_values(by='Rape Cases', ascending=False)
```

```python
topten=crime_by_state.head(10)
topten.plot(kind='bar', stacked=True)
plt.figure(figsize=(12,8))
sns.barplot(topten)
plt.xlabel("crime")
plt.title("crime in Top 10 States")
```
