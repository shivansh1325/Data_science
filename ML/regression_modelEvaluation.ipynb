```python
import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV,train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
data= pd.read_csv("../ML/boston_house_prices.csv",header= 1)
data.head()
```

```python
from sklearn.datasets import load_b
```

```python
x = data.drop(columns= ['MEDV'])
x.head()
```

```python
y=data['MEDV']
y.head()
```

```python
data.isnull().sum()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.35,random_state= 46)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
# defining the parameter gride for 
param_distribution={
    "n_estimators":[int(x) for x in np.linspace(100,1000, num= 10)],
    "max_features":["auto","sqrt","log"],
    "max_depth":[int(x) for x in np.linspace(10,110,num=11)],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4],
    "bootstrap":[True,False]
}
```

```python
for x in np.linspace(1,100,num=9):
    print(x)
```

```python
#setting randomizedSearchcv 
random_searchcv= RandomizedSearchCV(
    estimator=model,
    param_distributions=param_distribution,
    n_iter=200,
    cv=5,
    verbose=2,
    random_state= 45,
    n_jobs=-1
)
```

```python
random_searchcv.fit(x_train,y_train)
```

```python
#best hyperparameter
best_param = random_searchcv.best_estimator_
print(best_param)
y_pred = best_param.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
r2_ = r2_score(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print("root mean square error",rmse)
print("root mean square error",r2_)
print("root mean square error",mae)
```
